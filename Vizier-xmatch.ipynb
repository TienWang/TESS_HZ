```python
data = pd.read_csv('tic_v12.csv',index_col=0)
data.dropna(subset=['Teff'],inplace=True)
data.rename(columns={'lum':'Lum'},inplace=True)
z = (data.Lum == 0)
data.loc[z,'Lum']=np.NaN
```

```python

z = (data.Lum == 0)
data.loc[z,'Lum']=np.NaN
m = data.Lum.isna()
m += data.Kmag.isna()
nan = data[m].copy()
nan.drop(index=(nan.iloc[31].name),inplace=True)

v = Vizier(columns=['_RAJ2000', '_DEJ2000','Lum','Kmag'])
v.TIMEOUT=180
for num,i in enumerate(nan.index):
    print('Starting searching for object {}.'.format(str(num)))
    ra = nan.loc[i,'ra']
    dec = nan.loc[i,'dec']
    dic = {'Lum' : np.isnan(nan.loc[i,'Lum']),'Kmag':np.isnan(nan.loc[i,'Kmag'])}

    
    result = v.query_region(coord.SkyCoord(ra=ra, dec=dec,
                                               unit=(u.deg, u.deg),
                                               frame='icrs'),radius=5*u.arcsec)
    
        
    for prop,t_f in dic.items():
        if t_f:
            print('Searching for {}.'.format(prop))
            for cat in result.keys():
                if (prop in result[cat].colnames):
                    if result[cat][prop].compressed().shape[0]>0:
                        if not isinstance(result[cat][prop].compressed()[0],str):
                            if result[cat][prop].compressed().max()>0:
                                nan.loc[i,prop] = result[cat][prop].compressed().max()
                    
                    
    print('Ending searching for object {}.'.format(str(num)))
```

```python
data['Kmag'].fillna(nan['Kmag'],inplace=True)
data['Lum'].fillna(nan['Lum'],inplace=True)
data
```
